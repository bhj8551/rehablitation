#### 학생수에 따라 크기가 다른 강의실을 배치

```python
df = pd.read_csv('../data/enrolment_2.csv')
df
```

1. 80명 이상의 학생이 수강하는 과목은 “Auditorium”에서 진행됩니다.
2. 40명 이상, 80명 미만의 학생이 수강하는 과목은 “Large room”에서 진행됩니다.
3. 15명 이상, 40명 미만의 학생이 수강하는 과목은 “Medium room”에서 진행됩니다.
4. 5명 이상, 15명 미만의 학생이 수강하는 과목은 “Small room”에서 진행됩니다.
5. 폐강 등의 이유로 status가 “not allowed”인 수강생은 room assignment 또한 “not assigned”가 되어야 합니다.

```python
df['course name'].value_counts()[df['course name'].value_counts() >= 80]
```

```python
auditorium_condition = df['course name'].value_counts()[df['course name'].value_counts() >= 80].index.tolist()
auditorium_condition
```

```python
large_room_condition = df['course name'].value_counts()[(df['course name'].value_counts() >= 40) & (df['course name'].value_counts() < 80)].index.tolist()
large_room_condition
```

```python
medium_room_condition = df['course name'].value_counts()[(df['course name'].value_counts() >= 15) & (df['course name'].value_counts() < 40)].index.tolist()
medium_room_condition
```

```python
small_room_condition = df['course name'].value_counts()[(df['course name'].value_counts() >= 5) & (df['course name'].value_counts() < 15)].index.tolist()
small_room_condition
```

```python
df.loc[df['course name'].isin(auditorium_condition), 'room assignment'] = 'Auditorium'
```

```python
df.loc[df['course name'].isin(large_room_condition), 'room assignment'] = 'Large room'
```

```python
df.loc[df['course name'].isin(medium_room_condition), 'room assignment'] = 'Medium room'
```

```python
df.loc[df['course name'].isin(small_room_condition), 'room assignment'] = 'Small room'
```

```python
df.loc[df['status'] == 'not allowed', 'room assignment'] = 'not assigned'
```

#### 같은 크기의 강의실이 필요한 과목에 대해 알파벳 순서대로 방 번호를 배정

- 같은 크기의 강의실이 필요한 과목에 대해 알파벳 순서대로 방 번호를 배정하세요.

예를 들어 "Auditorium"이 필요한 과목으로 “arts”, “commerce”, “science” 세 과목이 있다면, “arts”는 “Auditorium-1”, “commerce”는 “Auditorium-2”, “science”는 “Auditorium-3” 순서로 방 배정이 되어야 합니다.

방 번호에 room 은 포함되지 않습니다. 아래 스크린샷을 참고하여 작성해주세요.

- status column이 “not allowed”인 수강생은 room assignment column을 그대로 “not assigned”로 남겨둡니다. "not allowed" 인 수강생의 room assignment 상태가 변경되지 않도록 유의해주세요.

- room assignment column의 이름을 room number로 바꿔주세요.

```python
df = pd.read_csv('../data/enrolment_3.csv')
df
```

```python
auditorium_courses = df.loc[df['room assignment'] == 'Auditorium', 'course name'].unique()
auditorium_courses
```

```python
auditorium_courses.sort()
auditorium_courses
```

```python
count = 1
for course in auditorium_courses:
    df.loc[df['course name'] == course , 'room number'] = 'Auditorium' + '-' + str(count)
    count += 1
```

```python
df
```

```python
large_room_courses = df.loc[df['room assignment'] == 'Large room', 'course name'].unique()
large_room_courses.sort()
count = 1
for course in large_room_courses:
    df.loc[df['course name'] == course , 'room number'] = 'Large' + '-' + str(count)
    count += 1
```

```python
medium_room_courses = df.loc[df['room assignment'] == 'Medium room', 'course name'].unique()
medium_room_courses.sort()
count = 1
for course in medium_room_courses:
    df.loc[df['course name'] == course , 'room number'] = 'Medium' + '-' + str(count)
    count += 1
```

```python
small_room_courses = df.loc[df['room assignment'] == 'Small room', 'course name'].unique()
small_room_courses.sort()
count = 1
for course in small_room_courses:
    df.loc[df['course name'] == course , 'room number'] = 'Small' + '-' + str(count)
    count += 1
```
